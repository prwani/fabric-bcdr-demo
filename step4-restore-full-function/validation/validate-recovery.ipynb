# Recovery validation notebook

Attach this notebook to the target lakehouse before running it. It validates that the expected sample tables exist and contain rows.

```python
EXPECTED_TABLES = ['dim_customer', 'dim_product', 'dim_region', 'dim_calendar', 'fact_sales']

table_rows = spark.sql('SHOW TABLES').collect()
available_tables = {row.tableName for row in table_rows}
missing_tables = [table for table in EXPECTED_TABLES if table not in available_tables]

if missing_tables:
    raise RuntimeError(f'Missing expected tables: {missing_tables}')

print('All expected lakehouse tables exist.')
```

```python
row_counts = {}
for table_name in EXPECTED_TABLES:
    row_count = spark.table(table_name).count()
    row_counts[table_name] = row_count
    if row_count <= 0:
        raise RuntimeError(f'Table {table_name} exists but has no rows')

row_counts
```

```python
total_rows = sum(row_counts.values())
print(f'Total validated rows: {total_rows:,}')
print('Lakehouse recovery validation passed.')
```
